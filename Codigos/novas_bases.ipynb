```python
dicionario = pd.read_excel("dicionario_PNS_microdados_2019_23062023.xlsx", header=1)
dicionario
```

```python
dic_limpo = dicionario.drop(dicionario.columns[[0, 1, 3, 5, 6]], axis=1)
dic_limpo
```

```python
dic_limpo = dic_limpo.rename(columns={
    dic_limpo.columns[0]: "codigo",
    dic_limpo.columns[1]: "descricao"
})
dic_limpo
```

```python
import pandas as pd

# Quantos são nulos na coluna 0 e na coluna 1
nulos_0_e_1 = dic_limpo[dic_limpo['codigo'].isna() & dic_limpo['descricao'].isna()].shape[0]

# Quantos não são nulos na coluna 0 e na coluna 1
nao_nulos_0_e_1 = dic_limpo[~dic_limpo['codigo'].isna() & ~dic_limpo['descricao'].isna()].shape[0]

# Quantos são nulos na coluna 0, mas não na coluna 1
nulos_0_nao_1 = dic_limpo[dic_limpo['codigo'].isna() & ~dic_limpo['descricao'].isna()].shape[0]

# Quantos são nulos na coluna 1, mas não na coluna 0
nulos_1_nao_0 = dic_limpo[~dic_limpo['codigo'].isna() & dic_limpo['descricao'].isna()].shape[0]

print("Nulos em 0 e 1:", nulos_0_e_1)
print("Não nulos em 0 e 1:", nao_nulos_0_e_1)
print("Nulos em 0, não em 1:", nulos_0_nao_1)
print("Nulos em 1, não em 0:", nulos_1_nao_0)
```

```python
# Remove linhas com código nulo
dic_limpo.dropna(subset=['codigo'], inplace=True)

# Reseta o índice
dic_limpo.reset_index(drop=True, inplace=True)

dic_limpo
```

```python
# Filtra dic_limpo para manter apenas códigos que estão nas colunas do df_filtrado
dic_limpo = dic_limpo[dic_limpo['codigo'].isin(df_filtrado.columns)].reset_index(drop=True)

dic_limpo
```

```python
# Ordenar dic_limpo na mesma ordem das colunas de df_filtrado
dic_limpo['codigo'] = pd.Categorical(dic_limpo['codigo'], categories=df_filtrado.columns, ordered=True)
dic_limpo = dic_limpo.sort_values('codigo').reset_index(drop=True)

# Salvar em CSV
dic_limpo.to_csv('dicionario_hip.csv', index=False, sep="|")
```
