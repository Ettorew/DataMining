```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from missforest import MissForest

# --- 1. PRÉ-PROCESSAMENTO (Como antes) ---
df_processado = df_hipertensao.copy()
categoricos_nomes = df_processado.select_dtypes(exclude=[np.number]).columns
mapeamentos_categoricos = {}

if not categoricos_nomes.empty:
    print(f"Codificando colunas: {categoricos_nomes.tolist()}")
    for col in categoricos_nomes:
        df_processado[col] = df_processado[col].astype('category')
        mapeamentos_categoricos[col] = dict(enumerate(df_processado[col].cat.categories))
        df_processado[col] = df_processado[col].cat.codes
        df_processado[col] = df_processado[col].replace(-1, np.nan)
    print("Codificação concluída.")
else:
    print("O DataFrame já era numérico.")

# --- 2. NOVO PASSO DE FORÇAR O TIPO ---
# Vamos forçar TUDO para float. Isso padroniza a entrada.
try:
    df_processado = df_processado.astype(float)
    print("\nSUCESSO: Todas as colunas foram forçadas para 'float'.")
except Exception as e:
    print(f"\nERRO ao forçar para float: {e}")
    print("Verifique os tipos de dados novamente com df_processado.info()")

# --- 3. DIVISÃO TREINO/TESTE ---
colunas_originais = df_processado.columns
train_df, test_df = train_test_split(df_processado, test_size=.3, shuffle=True,
                                   random_state=42)

# --- 4. IMPUTAÇÃO ---
# (Avisamos quais colunas tratar como categóricas)
# (Usamos 'categoricos_nomes' que pegamos lá no início)
imputer = MissForest(categorical=categoricos_nomes.tolist())

print("\nIniciando imputação nos dados de TREINO...")
train_imputed_array = imputer.fit_transform(x=train_df)
print("Treino imputado.")

print("Aplicando imputação nos dados de TESTE...")
test_imputed_array = imputer.transform(x=test_df)
print("Teste imputado.")

# --- 5. CONVERTER DE VOLTA PARA DATAFRAME ---
df_train_imputado = pd.DataFrame(train_imputed_array, columns=colunas_originais)
df_test_imputado = pd.DataFrame(test_imputed_array, columns=colunas_originais)

print("\n--- Head do DataFrame de Treino Imputado ---")
display(df_train_imputado.head())
```

```python
df_processado = pd.concat(
    [df_train_imputado, df_test_imputado], 
    ignore_index=True
)

display(df_processado.head())
```

```python
# --- 1. FAÇA UMA CÓPIA PARA REVERTER ---
df_revertido = df_processado.copy()

# Pega os dtypes de antes da imputação (ex: int8, float64)
tipos_originais_processados = df_processado.dtypes 

print("Iniciando reversão (decodificação) do DataFrame completo...")

# --- 2. LOOP DE REVERSÃO ---
for col_nome in df_revertido.columns:
    
    # Se esta coluna estava no nosso mapa de decodificação...
    if col_nome in mapeamentos_categoricos:
        
        mapeamento = mapeamentos_categoricos[col_nome]
        
        # Arredonda (ex: 0.99 -> 1), converte para int (1.0 -> 1)
        # e usa o mapa (1 -> 'Sim')
        df_revertido[col_nome] = df_revertido[col_nome].round(0).astype(int).map(mapeamento)
        
        # Converte de volta para 'category' para economizar memória
        df_revertido[col_nome] = df_revertido[col_nome].astype('category')

    # Se era originalmente um inteiro (mas não uma categoria)...
    elif 'int' in str(tipos_originais_processados.get(col_nome)).lower():
        
        # Arredonda e converte de float para int (ex: 58.0 -> 58)
        df_revertido[col_nome] = df_revertido[col_nome].round(0).astype(int)
        
    # Colunas float (como Peso) ficam como estão (float)

print("Processo concluído!")

# --- 3. VERIFIQUE O RESULTADO FINAL ---
print("\n--- DataFrame Completo, Imputado e Revertido ---")
display(df_revertido.head())

print("\n--- Tipos de Dados Finais ---")
df_revertido.info()
```
