```python
import pandas as pd
import numpy as np

df = pd.read_parquet("pns_filtrada.parquet").copy()
df = df.replace('NA', np.nan)
df
```

```python
df_hipertensao = df.loc[df['Q00201'] == '1'].copy()

def converter_numerico(data, coluna):
    data.loc[:, coluna] = pd.to_numeric(data[coluna], errors='coerce')

cols = [
    "P01101", "P013", "P015", "P023", "P00901",
    "P018", "P019", "P02001", "P02002", "P035", "P03701",
    "P029", "P03702", "Q003", "Q064"
]

for col in cols:
    if col in df_hipertensao.columns:
        converter_numerico(df_hipertensao, col)

df_hipertensao = df_hipertensao.reset_index(drop=True)
df_hipertensao
```
